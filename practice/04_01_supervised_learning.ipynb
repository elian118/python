```python
import numpy as np
import pandas as pd
```

```python
# 미리 seeds_dataset.csv 파일 배치 ← 데이터 출처: https://archive.ics.uci.edu/datasets
ysdata = pd.read_csv("./seeds_dataset.csv")
ysdata
```

```python
wheat_dataset = pd.DataFrame(ysdata)
wheat_dataset
```

```python
# 컬럼(행) 항목의 이름 나열
print(f"wheat_dataset의 키: {wheat_dataset.keys()}")

# 테스트 목적: 품종(breed) 외 다른 데이터를 분석해 품종(breed)을 알아낼 수 있는가?
```

```python
print(f"wheat_dataset의 크기: {wheat_dataset.shape}")
```

```python
print(f"wheat_dataset의 처음 다섯 행:\n{wheat_dataset[:5]}")
```

```python
datas = wheat_dataset.iloc[:, 0:7] # 밀의 특징 값만 분리 (breed 제외)
print(f"datas의 처음 다섯 행:\n{datas[:5]}")
```

```python
answer = wheat_dataset['breed']
print(f"answer 처음 다섯 행:\n{answer[:3]}")
print(f"answer 처음 다섯 행:\n{answer[107:110]}")
print(f"answer 처음 다섯 행:\n{answer[207:]}")
```

```python
from sklearn.model_selection import train_test_split

# 훈련용과 테스트용으로 데이터 분할 → 분할 목적: 미리 분할하지 않고 예측하면 이미 알고리즘이 답을 알고 있어 정확도가 항상 100%로 수렴(알고리즘 훈련 의미가 없음)
datas_train, datas_test, answer_train, answer_test = train_test_split(datas, answer, random_state=0)

print(f"datas_train 크기: {datas_train.shape}") # 전체 210개중 157개를 훈련용으로 분리
print(f"answer_train 크기: {answer_train.shape}")
```

```python
print(f"datas_test 크기: {datas_test.shape}")   # 전체 210개중 훈련용 데이터를 제외한 53개를 평가용 데이터로 분리
print(f"answer_test 크기: {answer_test.shape}")
```

```python
# knn(최근접 이웃)은 실무에서는 안 쓰는 아주 간단한 기본 저성능 알고리즘 모델. 알고리즘을 어떻게 구성하는지 이해하는 교보재로 활용됨
from sklearn.neighbors import KNeighborsClassifier # knn 에스티메이터(알고리즘이 구현된) 활용
knn = KNeighborsClassifier(n_neighbors=1) # 훈련할 객체 knn 생성 → n_neighbors: 주변 가장 가까이 있는 값. 숫자가 클수록 정확도가 크고 계산속도가 느려진다.
```

```python
knn.fit(datas_train, answer_train) # 생성된 훈련 모델 확인
```

```python
# 평가는 score 메서드로 확인
print("테스트 정확도: {:.2f}%".format(knn.score(datas_test, answer_test) * 100))
```

```python
# 208번 데이터 정답은 3
datas_new = np.array([[11.84, 13.21, 0.8521, 5.175, 2.836, 3.598, 5.044]]) # area  perimeter  compactness  length  width  asymmetry
print(f"X_new.shape: {datas_new.shape}")
# 실전에서 사용(predict)하면 측정값 넣었을 때 예측값이 나옴
prediction = knn.predict(datas_new)
print("prediction =", prediction) # 3 ← breed
print(f"prediction.shape: {prediction.shape}")
```

```python
# 테스트 데이터 53개 전체로 얻은 예측 결과들
answer_pred = knn.predict(datas_test)
print(f"테스트 셋에 대한 예측값:\n{answer_pred}")
```
